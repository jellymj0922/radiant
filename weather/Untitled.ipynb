```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os 
from tqdm import tqdm
from sklearn.model_selection import train_test_split
import xgboost as xgb
from xgboost import XGBRegressor
```

```python
path = "./uv_nooutlier/"
dir_list = os.listdir(path)
dir_train = dir_list
```

```python

data = pd.DataFrame()
for files in dir_train:
    df = pd.read_csv(path+files)
    X = df.drop(['uv'], axis=1)
    y= df['uv']
    
    from xgboost import XGBRegressor
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)
    xgb = XGBRegressor(n_estimators = 100, learning_rate = 0.017, max_depth=11, subsample=0.5,colsample_bytree=0.7,  random_state = 2000).fit(X_train,y_train, eval_set=[(X_test, y_test)], 
    early_stopping_rounds=15)
    
    

    
    print("------------------------------------")
    print("files")
    
    df_imp = pd.DataFrame({'imp':xgb.feature_importances_}, index = xgb.get_booster().feature_names)
    df_imp = df_imp[df_imp.imp > 0].sort_values('imp').copy()

    feat_num = df_imp.shape[0]
    print("total number of features =", feat_num)
    print(df_imp)
```
