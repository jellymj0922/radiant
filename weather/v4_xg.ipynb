```python
X_train = train.drop(['YearMonthDayHourMinute', 'STN', 'Lon', 'Lat', 'isitu-LST',
       '30daysBand3', '30daysBand13', 'GK2A-LST',  'insitu-TED0.1', 'insitu-TED0.2',
       'insitu-TED0.3', 'insitu-TED0.5', 'insitu-TED1.0', 'insitu-TED1.5',
       'insitu-TED3.0', 'insitu-TED5.0', 'insitu-PA', 'insitu-PS', 'Year',
       'Month', 'Day', 'Hour', 'Minute',  'mean_34_37',
        'mean_43_44','Band2','Band3','Band4','Band5','Band6','Band7','Band8','Band9','Band10','Band11','Band12',
            'Band13','Band14','Band15','Band16','30daysBand3','30daysBand13','insitu-TA'], axis=1)
X_test = val.drop(['YearMonthDayHourMinute', 'STN', 'Lon', 'Lat', 'isitu-LST',
       '30daysBand3', '30daysBand13', 'GK2A-LST',  'insitu-TED0.1', 'insitu-TED0.2',
       'insitu-TED0.3', 'insitu-TED0.5', 'insitu-TED1.0', 'insitu-TED1.5',
       'insitu-TED3.0', 'insitu-TED5.0', 'insitu-PA', 'insitu-PS', 'Year',
       'Month', 'Day', 'Hour', 'Minute',  'mean_34_37',
       'mean_43_44','Band2','Band3','Band4','Band5','Band6','Band7','Band8','Band9','Band10','Band11','Band12',
            'Band13','Band14','Band15','Band16','30daysBand3','30daysBand13','insitu-TA'], axis=1)

y_train=train[['isitu-LST', 'insitu-TA']].values
y_test=val[['isitu-LST', 'insitu-TA']].values

X_train
```

```python
import xgboost as xgb
# xg_reg = xgb.XGBRegressor(objective ='reg:linear', colsample_bytree = 0.3, learning_rate = 0.1, max_depth = 5, alpha = 10, n_estimators = 10)
# xg_reg.fit(X_train,y_train)

xgb_modelv4 = xgb.XGBRegressor(n_estimators = 200, learning_rate = 0.1, random_state = 2000).fit(X_train,y_train, eval_set=[(X_test, y_test)], verbose=False)
```

```python
xgb_modelv4
```

```python
filename = 'xgb_modelv4.model'
xgb_modelv4.save_model(filename) #save
```

```python
# 모델 불러오기
new_xgb_modelv4 = xgb.XGBRegressor() # 모델 초기화
new_xgb_modelv4.load_model(filename) # 모델 불러오기
```

```python
from sklearn.metrics import mean_squared_error

preds =new_xgb_modelv4.predict(X_test)
mse = mean_squared_error(y_test, preds)
print("MSE: %f" % (mse))
```

```python
import matplotlib.pyplot as plt

lst_pr = []
ta_pr = []
for i in range(len(preds)):
    lst_pr.append(preds[i][0])
    ta_pr.append(preds[i][1])
    
lst_re = []
ta_re = []
for i in range(len(y_test)):
    lst_re.append(y_test[i][0])
    ta_re.append(y_test[i][1])
    
def value_compare():
  plt.subplot(211)
  plt.plot(lst_pr[500:550], color = 'b', label = 'pred')
  plt.plot(lst_re[500:550], color = 'g', label = 'real')
  plt.ylabel('LST')
  plt.legend()

  plt.subplot(212)
  plt.plot(ta_pr[700:750], color = 'r', label = 'pred')
  plt.plot(ta_re[700:750], color = 'black', label = 'real')
  plt.ylabel('TA')
  plt.legend()
  plt.show()  

value_compare()
```
